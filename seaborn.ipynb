# Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#shows the plots inside the notebook
%matplotlib inline
#autoreload
%reload_ext autoreload 
%autoreload 2
```

      Import Data

```python
#using seaborn built in datasets
print(sns.get_dataset_names())

#loading dataset
crash_df =sns.load_dataset('car_crashes')
crash_df.head(5)
```

    Distribution plots

```python
# used for univariate distribution
sns.distplot(crash_df['not_distracted'],kde=False ,bins =25) #bins are used to divide the data into buckets so as to see reduce the number of bars
```

        Joint plot(used to compare two plots)

```python
#plots a scatter plot by default

sns.jointplot(x='speeding' , y='alcohol', data=crash_df,kind='hex') # the kind instance is used to determine the kind of plot you want
plt.suptitle('Hexagen plot')
```

    KDE(Kernel Density estimate) PLOTS

```python
sns.kdeplot(crash_df['alcohol'])
```

    Pair plots

```python
#To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function.
sns.pairplot(crash_df)
```

```python
#loading a new dataframe
tips_df = sns.load_dataset('tips')
tips_df.head(5)
```

```python
sns.pairplot(tips_df,hue='sex',palette='Blues') #hue is used to pass in categorical data to which the analysis will be based on

```

        Rugplots

```python
sns.rugplot(tips_df['tip']) # It is used to visualise the distribution of the data and shows density. Not used often

```

    Styling

```python
#styling for the axis and the grids
sns.set_style('ticks')
#seting the size of the plot
plt.figure(figsize=(8,4))

sns.set_context('paper',font_scale=1.7)

sns.jointplot(x ='speeding',y='alcohol',data=crash_df,kind='reg')

#to remove the axis lines since the may not give any necessary information
# sns.despine(left =True ,bottom =True  )
```

# Categorical Plots

    Barplots

```python
sns.barplot(x ='sex' , y = 'total_bill',data=tips_df ,estimator=np.var) #by default the estimator is based on the mean

```

    Count plot

```python
#counts the no of occurrences
sns.countplot(x ='sex',data = tips_df)
 
```

    Boxplot

```python
sns.boxplot(x='day',y='total_bill', data = tips_df,hue='sex')

#to move the legend
plt.legend(loc = 0)
```

    Violin Plot

```python
#similar to boxplots and uses KDE's
sns.violinplot(x='day', y = 'total_bill',data =tips_df,hue='sex',split=True,)#split is used to compare the data to each other and uses KDEs
```

    Strip plots

```python
plt.figure(figsize = (8,5))

#this is used with a categorical variable and  is used similar to a boxplot
#dodge used to seperate the hue data
sns.stripplot(x='day',y='total_bill',data =tips_df,jitter=True,hue='sex',dodge=True) #jitter is used to spread out the data and avoid it being stacked to each other


```

    Swarm plot

```python
#used together with other plots and is similar to strip plots
sns.violinplot(x='day',y='total_bill',data=tips_df,)
sns.swarmplot(x='day',y='total_bill',data=tips_df,color ='white')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips_df,color ='black')
```

    Palettes

```python
plt.figure(figsize=(9,6))
sns.set_style('white')
sns.set_context('talk')
sns.stripplot(x='day',y='total_bill',data=tips_df,hue ='sex',palette='afmhot') #palette used to chnge color

## Fore more colors for the palette visit https://matplotlib.org/stable/tutorials/colors/colormaps.html
#upper right = 1, upper left =2, lower left =3, lower right =4
plt.legend(loc =2)

# for more legend info on the legend https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html

```

    Pair grids

```python
#similar to pair plots bit have precise control over the grids

iris =sns.load_dataset('iris')
iris_g = sns.PairGrid(iris,hue='species')

#iris_g.map(plt.scatter)#used to show that we use only scatter plots in the pair grid
iris_g.map_diag(plt.hist)
#Alternatively,
iris_g.map_offdiag(plt.scatter)#to determine what the off diagonal plots will be


```

```python
iris =sns.load_dataset('iris')
iris_g = sns.PairGrid(iris,hue='species')

iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
iris_g.map_diag(plt.hist)
```

```python

#defining variables for the plots
iris_h = sns.PairGrid(iris,hue ='species',x_vars=['sepal_length','sepal_width'],
                      y_vars=['petal_length','petal_width'])
iris_h.map(plt.scatter)
iris_h.add_legend()
```

    Facet Grid

```python
tips_fg = sns.FacetGrid(tips_df,col ='time',row ='smoker')
#tips_fg.map(plt.hist,'total_bill',bins =8)
tips_fg.map(plt.scatter,'total_bill','tip')
```

```python
tips_fg2 = sns.FacetGrid(tips_df,col ='time',hue ='smoker',height= 4, aspect=1.2, 
                         col_order=['Dinner','Lunch'],palette='Set1')
tips_fg2.map(plt.scatter,'total_bill','tip',edgecolor='w')
```

```python
kws =dict(s=50, linewidth = 0.5,edgecolor ='w')
tips_fig2 =sns.FacetGrid(tips_df,col ='sex',hue ='smoker',height= 4 ,aspect=1.2,
                         hue_order=['Yes','No'],hue_kws=dict(marker=['^','v']))
tips_fig2.map(plt.scatter,'total_bill','tip',**kws)
```
